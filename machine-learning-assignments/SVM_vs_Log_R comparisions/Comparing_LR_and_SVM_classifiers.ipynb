```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[1]*mi - intercept)/coef[0]), mi],[((-coef[1]*ma - intercept)/coef[0]), ma]])
    return points
    #plt.plot(points[:,0], points[:,1])
```

```python
# here we are creating 2d imbalanced data points and applying LinearSVC
ratios = [(100,2), (100, 20), (100, 40), (100, 80)]
plt.figure(figsize=(23,15))


for j,i in enumerate(ratios):
  for m in range(3):
    plt.subplot(4,3,3*j+m+1)
    X_p=np.random.normal(0,0.05,size=(i[0],2))
    X_n=np.random.normal(0.13,0.02,size=(i[1],2))
    y_p=np.array([1]*i[0]).reshape(-1,1)
    y_n=np.array([0]*i[1]).reshape(-1,1)
    X=np.vstack((X_p,X_n))
    y=np.vstack((y_p,y_n))
   

    mean_X = X.mean(axis=0)[0]
    std_dev_X = X.std(axis=0)[0]
    mean_X_d2 =  X.mean(axis=0)[1]
    std_dev_X_d2 = X.std(axis=0)[0]

    svm_clf = Pipeline([
                    ("Scaler", StandardScaler()),
                    ("linear_svc", LinearSVC(C=1, loss="hinge")) \
                    ])
    svm_clf.fit(X,y)
    #print(svm_clf.predict([[0.10,0.10]]))

    coef , inter = svm_clf[1].coef_[0], svm_clf[1].intercept_[0]
    y_min, y_max = -2, +2 #these can be chosen other values also
    
    points = draw_line(coef, inter, y_min, y_max)
    #print(points)
    plt.plot(points[:,0]*std_dev_X + mean_X, points[:,1]*std_dev_X_d2 + mean_X_d2)
    plt.scatter(X_p[:,0],X_p[:,1])
    plt.scatter(X_n[:,0],X_n[:,1],color='red')
plt.show()
```

```python
# here we are creating 2d imbalanced data points and applying Logistic Regression
ratios = [(100,2), (100, 20), (100, 40), (100, 80)]
plt.figure(figsize=(23,15))


for j,i in enumerate(ratios):
  for m in range(3):
    plt.subplot(4,3,3*j+m+1)
    X_p=np.random.normal(0,0.05,size=(i[0],2))
    X_n=np.random.normal(0.13,0.02,size=(i[1],2))
    y_p=np.array([1]*i[0]).reshape(-1,1)
    y_n=np.array([0]*i[1]).reshape(-1,1)
    X=np.vstack((X_p,X_n))
    y=np.vstack((y_p,y_n))

    mean_X = X.mean(axis=0)[0]
    std_dev_X = X.std(axis=0)[0]
    mean_X_d2 =  X.mean(axis=0)[1]
    std_dev_X_d2 = X.std(axis=0)[0]

    svm_clf = Pipeline([
                    ("Scaler", StandardScaler()),
                    ("linear_svc", LogisticRegression()) \
                    ])
    svm_clf.fit(X,y)

    coef , inter = svm_clf[1].coef_[0], svm_clf[1].intercept_[0]
    y_min, y_max = -2, +2 #these can be chosen other values also
    
    points = draw_line(coef, inter, y_min, y_max)
    plt.plot(points[:,0]*std_dev_X + mean_X, points[:,1]*std_dev_X_d2 + mean_X_d2)
    plt.scatter(X_p[:,0],X_p[:,1])
    plt.scatter(X_n[:,0],X_n[:,1],color='red')
plt.show()
```
